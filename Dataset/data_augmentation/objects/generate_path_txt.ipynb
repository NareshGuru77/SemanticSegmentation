```python
import os
import os.path as ph
import shutil
LABEL_PATH = './label'
IMAGE_PATH = './image'
import numpy as np
```

```python
object_paths = list()
for obj_dir in os.listdir(IMAGE_PATH):
    obj_files = list()
    for files in sorted(os.listdir(os.path.join(IMAGE_PATH, obj_dir))):
        obj_files.append([os.path.join(IMAGE_PATH.split('/')[1], obj_dir, files), 
                         os.path.join(LABEL_PATH.split('/')[1], obj_dir, files.split('.')[0]+'.png')])
    object_paths += obj_files.copy()
```

```python
to_write = np.array(object_paths)
to_write
```

```python
np.savetxt('./atWork_train_list.txt', to_write, fmt="%s")
```
